```python
import pandas as pd
import numpy as np
import random
import string
np.random.seed(56)
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
def score():
    return np.random.randint(0, 100)
```

```python
df = pd.DataFrame([[score() for i in range(1, 5)] for j in range(1, 11)],
             index=[id() for i in range(1, 11)], 
             columns=['국어', '영어','수학', '사회']) 
```

```python
df.loc[:, '국어'] # 특정 컬럼 모든 행 조회
```

```python
df.loc[['bwePv'],:]  # 특정 행 모든 컬럼 조회 방법 1
```

```python
df.iloc[[0],:]      # 특정 행 모든 컬럼 조회 방법 2
```

```python
# 과학이 추가되었습니다. 기존 학생들에게 각 랜덤으로 추가하시오
df['과학'] = np.random.randint(1, 100) for i in range(1,11)
df
```

```python
df.loc[:,'과학'] = pd.Series([score()  for i in range(1,11)], 
                           index=df.index)
df
```

```python
# 각 학생들의 점수의 총합을 표현하는 컬럼을 추가하세요.
df['총점'] = df.sum(axis=1)
df
```

```python
# 각 학생들의 점수의 총합을 표현하는 행을 추가하세요.
df.loc['합계'] = df.sum(axis=0)
df
```

```python
# 총점기준으로 1등부터 10등까지
df['등수'] = df['총점'].rank(axis=0, method='max', ascending = True)
```
